```python
df = pd.read_csv("Demo.csv")
```

```python
df.info()
df.describe()
```

```python
df.isnull().sum()
```

```python
# df[df['UT 1 Marks'].isnull()]
df[df['%.1'].isnull()]
# df[df['%'].isnull()]
# df[df['50% maks of Previous Year SPPU Result'].isnull()]



```

```python
df['UT 1 Marks'].fillna(df['UT 1 Marks'].mean(), inplace=True)
```

```python
df['%.1'].fillna(df['%.1'].median(), inplace=True)
```

```python
df['%'].fillna(df['%'].mode()[0], inplace=True)
```

```python
df['50% maks of Previous Year SPPU Result'].fillna(df['50% maks of Previous Year SPPU Result'].mode()[0], inplace=True)
```

```python
df.head()
```

```python
numeric_cols = df.select_dtypes(include=['int64','float64']).columns
numeric_cols
```

```python
import matplotlib.pyplot as plt

for col in numeric_cols:
    plt.figure()
    df.boxplot(column=col)
    plt.title(f"Boxplot of {col}")
    plt.show()
```

```python
sns.scatterplot(x='UT 1 Marks', y='50% maks of Previous Year SPPU Result', data=df)
plt.show()
```

```python
import itertools

for col1, col2 in itertools.combinations(numeric_cols, 2):
    plt.figure()
    sns.scatterplot(x=df[col1], y=df[col2])
    plt.xlabel(col1)
    plt.ylabel(col2)
    plt.title(f"{col1} vs {col2}")
    plt.show()
```

```python
import numpy as np
# Creating a new column with the base-10 logarithm of the scores
df['Log_Marks'] = np.log10(df['UT 1 Marks'])
```

```python
df['Log_Marks'].plot(kind='hist', title='Log-Transformed UT 1 Marks')
plt.show()
```
